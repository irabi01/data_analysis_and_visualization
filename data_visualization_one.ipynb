```python
days = [1,2,3,4,5,]
sleeping = [7,5,7,10,5]
playing = [5,7,9,2,7]
eating = [7,3,7,16,5]
watching = [2,10,8,19,10]
plt.plot([],[], color = 'm', label = 'Sleeping', linewidth = 5)
plt.plot([],[], color = 'c', label = 'Playing', linewidth = 5)
plt.plot([],[], color = 'r', label = 'Eating', linewidth = 5)
plt.plot([],[], color = 'k', label = 'watching', linewidth = 5)

plt.stackplot(days, sleeping, playing, eating, watching, colors = ['m', 'c', 'r', 'k'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Interesting graph \n check it out')
plt.legend()
plt.show()
```

```python
cols = ['m','c','r','b']
slices = [7,5,2,9]# all labels depending on slice
activities = ['sleeping', 'playing', 'eating', 'watching']
plt.pie(slices, labels = activities, colors = cols, startangle = 90, shadow = True, explode = (0,0.1,0,0), autopct = '% 1.1f %%')
plt.title('Interesting graph \n check it out')
plt.show()
```
